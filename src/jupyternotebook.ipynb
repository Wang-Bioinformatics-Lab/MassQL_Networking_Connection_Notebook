```python
import pandas as pd
from gnpsdata import workflow_massql
from gnpsdata import workflow_classicnetworking
from gnpsdata import taskresult
```

```python
# This workflow
```

```python
networking_task = "05609e85f5bb4ee9a252e62e275586d2"
falcon_task = "9a24bc9c215547adadbf8dcde4a74053"
massql_task = "51d9df72575b4e07bd23aa86cd752b4d"

molecular_component_of_interest = 10
```

```python
# Getting the nodes for this component
nodes_df = workflow_classicnetworking.get_clustersummary_dataframe(networking_task)
nodes_df = nodes_df[nodes_df["componentindex"] == molecular_component_of_interest]
nodes_df = nodes_df[["cluster index"]]

all_nodes = set(nodes_df["cluster index"])

# Getting cluster info from networking
clusterinfo_df = workflow_classicnetworking.get_clusterinfo_dataframe(networking_task)
nodes_clusterinfo_df = clusterinfo_df[clusterinfo_df["cluster index"].isin(all_nodes)]
```

```python
# This finds all scans that were in the clustered falcon spectra
falcon_clustermapping_df = taskresult.get_task_resultview_dataframe(falcon_task, "view_clusterinfo")

all_falcon_scans = set(nodes_clusterinfo_df["cluster index"])
falcon_clustermapping_df = falcon_clustermapping_df[falcon_clustermapping_df["cluster"].isin(all_falcon_scans)]
```

```python
# Getting MassQL Mapping for Extraction
massql_extraction_df = workflow_massql.get_extractionresults_dataframe(massql_task)

falcon_clustermapping_df["key"] = falcon_clustermapping_df["filename"] + falcon_clustermapping_df["scan"].astype(str)
massql_extraction_df["key"] = massql_extraction_df["new_filename"] + massql_extraction_df["new_scan"].astype(str)

# Filtering
massql_extraction_df = massql_extraction_df[massql_extraction_df["key"].isin(falcon_clustermapping_df["key"])]

massql_extraction_df.to_csv("MassQL_scans_in_component.tsv", sep="\t")
```
